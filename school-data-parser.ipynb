```python
import pandas as pd
import xlrd
```

```python
# read in all of the relevant data sets
# 1. High School Grad Plans (started with this one so only high schools are listed)
# 2. SATs by School
# 3. Discipline by School (get school size)
# 4. School Directory (contact, location, and district info)
df = pd.read_excel("Cleartelligence Data Visualization Evaluation Data.xlsx",sheet_name="High School Grad plans by Schoo")
sats = pd.read_excel("Cleartelligence Data Visualization Evaluation Data.xlsx",sheet_name="SATs by School")
discipline = pd.read_excel("Cleartelligence Data Visualization Evaluation Data.xlsx",sheet_name="Discipline by School")
directory = pd.read_excel("Cleartelligence Data Visualization Evaluation Data.xlsx",sheet_name="School Directory")
```

```python
df = df.merge(sats,how="left")
df = df.merge(discipline,how="left")
df = df.merge(directory,how="left")
```
